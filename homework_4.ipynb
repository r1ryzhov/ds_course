```python
import numpy as np
```

```python
# level 1.1

a = np.array([15, 32, 0])
b = np.array([-32, -18, 25])

np.dot(a, b)
```

```python
# level 1.2
# 1
A = np.array([[5, 10, 1],
              [0, 1, 0],
              [9, 3, 6]])

# a)
B1 = np.array([
    [1, 7, 3],
    [0, 2, 0],
    [9, 4, 2]
])

print(f'a: {A @ B1}')

# b)
B2 = np.array([
    [5, 1, 8],
    [2, 6, 3],
    [1, 3, 0]
])

print(f'b: {A @ B2}')
```

```python
# 2
A = np.array([
    [5, 1, 8],
    [2, 6, 3],
    [1, 3, 0]
])

print(A.T)
```

```python
# 3

print(np.linalg.inv(A))
```

```python
# level 2
robo_cats = np.array([[100, 200, 1000],
                     [350, 150, 55],
                     [30, 30, 30]])

expenditures = np.array([2600, 1660, 210])

print(np.linalg.solve(robo_cats, expenditures), 'лапок, мордочек и хвостиков соответственно.')
```
